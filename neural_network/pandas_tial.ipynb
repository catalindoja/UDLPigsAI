```python
import pandas as pd
import json
import glob
import os


def readFiles(directory):
    df = pd.DataFrame()
    for filename in os.listdir(directory):
        file = os.path.join(directory, filename)
        data = json.load(open(file))
        df = df.append(data, ignore_index=True)

    return df



df = readFiles('./jsons/')
print(df)
```

```python
images_arr = []
annotations_arr = []
for img in images:
    temp = str(img).replace("[{", "").replace("}]", "")
    temp = temp.split(", ")
    images_arr.append(temp)
    
for annotation in annotations:
    temp = str(annotation).replace("[{", "").replace("}]", "")
    temp = temp.split(", ")
    annotations_arr.append(temp)
    
```
